```python
dicionario = {
    "Natureza":"Fim Lucrativo",
    "%": "Cursos Sem Ato 5 anos",
    "CI N": "Ultimo CI",
    "CI V": "Penultimo CI",
    "IGC_N": "Ultimo IGC",
    "IGC_M": "Penultimo IGC",
    "IGC_V": "Antepenultimo IGC",
    "variacao mat": "Variacao Matricula 16/17",
    "mat_T 2017": "Matriculas 17",
    "saldo 2017": "Saldo 2017",
    "Variacao 17-16": "Variacao do Saldo 16/17"
}

dados = dados.rename(columns=dicionario)
```

```python
dados.head(10)
```

```python
dados.info()
```

```python
dados["Ultimo CI"] = pd.to_numeric(dados["Ultimo CI"], errors='coerce')
dados["Ultimo CI"].replace(0,np.nan,inplace=True)
dados.info()
```

```python
dados["Ultimo CI"].describe()
```

```python
import matplotlib.pyplot as plt

dados_filtrados = dados

contagem_por_CI = dados_filtrados["Ultimo CI"].value_counts().sort_index()

plt.figure(figsize=(10, 6))

contagem_por_CI.plot(kind="bar", color="skyblue")
plt.xlabel('Ultimo CI')
plt.ylabel('Quantidade de IES')
plt.title('Quantidade de IES por Ultimo CI')


plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python

dados["Penultimo CI"] = pd.to_numeric(dados["Penultimo CI"], errors='coerce')
dados["Penultimo CI"].replace(0,np.nan,inplace=True)

dados_filtrados = dados

contagem_por_CI = dados_filtrados["Penultimo CI"].value_counts().sort_index()

plt.figure(figsize=(10, 6))

contagem_por_CI.plot(kind="bar", color="skyblue")
plt.xlabel('Penultimo CI')
plt.ylabel('Quantidade de IES')
plt.title('Quantidade de IES por Penultimo CI')

plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python

dados["Fim Lucrativo"] = dados["Fim Lucrativo"].map({1: "Lucrativo", 0: "No Lucrativo"})
dados["Saldo 2017 Positivo"] = dados["Saldo 2017"] > 0

dados_agrupados = dados.groupby(["Fim Lucrativo", "Saldo 2017 Positivo"]).size().unstack().fillna(0)

plt.figure(figsize=(10, 6))

dados_agrupados.plot(kind="bar", stacked=True, color=[ "lightcoral","skyblue"])
plt.xlabel('Fim Lucrativo')
plt.ylabel('Quantidade de Instituies')
plt.title('Relao entre Fim Lucrativo e Saldo 2017')
plt.legend(["Saldo 2017 Negativo", "Saldo 2017 Positivo"], loc="upper right")


plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
faixas = [-float('inf'), 3, 4, float('inf')]

categorias = ['Muito Insatisfatrio', 'Adequado', 'Excelente']

dados["Ultimo IGC"] = pd.to_numeric(dados['Ultimo IGC'], errors='coerce')

dados['Ultimo IGC Categoria'] = pd.cut(dados["Ultimo IGC"], bins=faixas, labels=categorias)

dados.head()
```

```python
dados["DIAS"] = dados["DIAS"].astype(int)
dados["Cursos Sem Ato 5 anos"] = pd.to_numeric(dados["Cursos Sem Ato 5 anos"], errors="coerce")
dados["Matriculas 17"] = dados["Matriculas 17"].astype(int)
dados["Variacao Matricula 16/17"] = pd.to_numeric(dados["Variacao Matricula 16/17"],errors="coerce")
dados["% FIES"] = pd.to_numeric(dados["% FIES"], errors="coerce")
dados["Saldo 2017"] = pd.to_numeric(dados["Saldo 2017"], errors="coerce")
dados["Variacao do Saldo 16/17"] = pd.to_numeric(dados["Variacao do Saldo 16/17"], errors="coerce")
dados["Penultimo IGC"] = pd.to_numeric(dados["Penultimo IGC"], errors="coerce")
dados["Antepenultimo IGC"] = pd.to_numeric(dados["Antepenultimo IGC"], errors="coerce")
dados["Fim Lucrativo"] = dados["Fim Lucrativo"].astype("category")
dados["Situacao"] = dados["Situacao"].astype("category")

dados.info()
```

```python

contagem_por_igc = dados["Ultimo IGC Categoria"].value_counts().sort_index()

plt.figure(figsize=(12, 6))

contagem_por_igc.plot(kind="bar", color="skyblue")
plt.xlabel('Ultimo IGC')
plt.ylabel('Quantidade de IES')
plt.title('Quantidade de IES por Ultimo IGC')

plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
contagem_situacao = dados_preprocessados['Situacao'].value_counts()

plt.figure(figsize=(8, 6))
ax = contagem_situacao.plot(kind='bar', color='skyblue')
plt.xlabel('Situao')
plt.ylabel('Quantidade')
plt.title('Quantidade por Situao')

# Adicione os valores nas barras
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom')

plt.xticks(rotation=45)  # Rotao dos rtulos no eixo x para melhor legibilidade
plt.show()
```

```python
# Suponha que seu DataFrame seja df

# Escolha as colunas numricas que voc deseja comparar em relao  coluna "Situacao"
colunas_numericas = [
    'DIAS',
    'Cursos Sem Ato 5 anos',
    'Ultimo CI',
    'Ultimo IGC',
    'Variacao Matricula 16/17',
    'Matriculas 17',
    '% FIES',
    'CURSOS',
    'Saldo 2017',
    'Variacao do Saldo 16/17',
]

# Crie subplots para a mdia e boxplots
fig, axes = plt.subplots(nrows=len(colunas_numericas), ncols=2, figsize=(10, 40))
plt.subplots_adjust(wspace=0.5,hspace=0.5)

# Loop atravs das colunas numricas
for i, coluna in enumerate(colunas_numericas):
    # Grfico de barras para a mdia
    sns.barplot(x='Situacao', y=coluna, data=dados_preprocessados, ax=axes[i-1, 0], errorbar='sd')
    axes[i-1,0].set_xlabel('')
    axes[i-1,0].set_ylabel('')
    axes[i-1,0].set_title(f'{coluna}')

    # Boxplot
    sns.boxplot(x='Situacao', y=coluna, data=dados_preprocessados, ax=axes[i-1, 1])
    axes[i -1, 1].set_xlabel('')
    axes[i -1, 1].set_ylabel('')
    axes[i -1, 1].set_title(f'{coluna}')

    # Ajuste os ttulos dos subplots
    axes[i-1, 0].set_ylabel('Valor')

    # Ajuste os rtulos das categorias da Situacao
    axes[i-1, 1].set_xlabel('Situacao')
    axes[1, 0].set_xlabel('Situacao')

# Defina os ttulos gerais
fig.suptitle('Comparao de Mdia e Boxplots por Situacao', fontsize=12)
plt.show()
```

```python
ativa = dados_preprocessados[dados_preprocessados['Situacao'] == 'Ativa']

# Filtrar os dados para situao extinta
extinta = dados_preprocessados[dados_preprocessados['Situacao'] == 'Extinta']

# Criar dois grficos de histograma
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.histplot(data=ativa, x='Ultimo CI', bins=range(1, 6), element='step', color='blue', discrete=True)
plt.xlabel('Ultimo CI')
plt.ylabel('Frequncia')
plt.title('Histograma de Ultimo CI para Situao Ativa')

# Histograma para situao extinta
plt.subplot(1, 2, 2)
sns.histplot(data=extinta, x='Ultimo CI', bins=range(1, 6), element='step', color='red', discrete=True)
plt.xlabel('Ultimo CI')
plt.ylabel('Frequncia')
plt.title('Histograma de Ultimo CI para Situao Extinta')

plt.tight_layout()
plt.show()
```
