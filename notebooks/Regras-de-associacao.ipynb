```python
dados['Fim Lucrativo'] = dados['Fim Lucrativo'].map({'Lucrativo': 1, 'Não Lucrativo': 0})
dados['Situacao'] = dados['Situacao'].map({'Ativa': 1, 'Extinta': 0})

dados.head()
```

```python
dados.info()
```

```python
dados.describe()
```

```python
# Definindo os intervalos para categorias
intervalos = [0, 885, 2421, 5331, float('inf')]  # Infinito é representado como float('inf')

# Definindo os rótulos para as categorias
rotulos = ['Até 885 dias', 'De 886 a 2421 dias', 'De 2421 até 5331 dias', 'Mais de 5332 dias']

# Usando pd.cut() para dividir a coluna 'dias' em categorias binárias
dados['categorias'] = pd.cut(dados['DIAS'], bins=intervalos, labels=rotulos, right=False)

# Convertendo as categorias em colunas binárias usando pd.get_dummies()
colunas_binarias = pd.get_dummies(dados['categorias'])

# Concatenando as colunas binárias ao DataFrame original
dados = pd.concat([dados, colunas_binarias], axis=1)

dados.head()
```

```python
dados = dados.drop(columns=["categorias"])
dados.head()
```

# Regras de Associação

```python
from mlxtend.frequent_patterns import apriori

itemsets = apriori(dados, min_support=0.5, use_colnames=True)

itemsets
```

```python
from mlxtend.frequent_patterns import association_rules

regras = association_rules(itemsets, metric="confidence", min_threshold=0.8)

regras.sort_values(by="lift", ascending=False)
```

```python
# Lista das colunas que você deseja contar e plotar
colunas_dias = ['Até 885 dias', 'De 886 a 2421 dias', 'De 2421 até 5331 dias', 'Mais de 5332 dias']

# Filtrar o DataFrame com base no critério
dados_dias_ativa = dados[dados['Situacao'] == 1]
dados_dias_extinta = dados[dados['Situacao'] == 0]

# Contar a quantidade de 1 em cada coluna para ambos os critérios
contagem_dias_ativas = dados_dias_ativa[colunas_dias].sum()
contagem_dias_extintas = dados_dias_extinta[colunas_dias].sum()

# Criar um gráfico de colunas com cores diferentes para diferentes critérios
largura_barra = 0.35
indices = np.arange(len(contagem_dias_ativas))

plt.figure(figsize=(8, 6))
plt.bar(indices, contagem_dias_ativas, largura_barra, color='skyblue', label='Ativa')
plt.bar(indices + largura_barra, contagem_dias_extintas, largura_barra, color='red', label='Extinta')
plt.title('Dias')
plt.xlabel('Colunas')
plt.ylabel('Quantidade')
plt.legend()
plt.xticks(indices + largura_barra / 2, contagem_dias_ativas.index, rotation=45)
plt.show()
```
